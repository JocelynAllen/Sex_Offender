```python
##Import statements
#http://opendata.dc.gov/datasets/sex-offender-registry?geometry=-77.416%2C38.81%2C-76.479%2C38.997
import pandas as pd
from sqlalchemy import create_engine
```

### Store Sex Offender CSV into DataFrame###

```python
csv_file = "Sex_Offender_Registry.csv"
offender_data_df = pd.read_csv(csv_file, encoding = 'latin1')
offender_data_df.head()
```

```python
connection_string = "root:#Mustang88#@localhost/sex_offender"
engine = create_engine(f'mysql://{connection_string}', pool_size=10, max_overflow=20,)
engine.table_names()
offender_data_df.to_sql(name='profile', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from profile', con=engine).head()
```
